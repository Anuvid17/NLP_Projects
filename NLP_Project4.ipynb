```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
```

```python
df = pd.read_csv("emails.csv")

df = df[['text']]

df.dropna(inplace=True)
```

```python
stop_words = set(stopwords.words('english'))
```

```python
spam_words = {

    'free',
    'offer',
    'winner',
    'win',
    'claim',
    'bonus',
    'cash',
    'click',
    'reward',
    'urgent',
    'limited',
    'congratulations',
    'prize',
    'money',
    'discount',
    'deal',
    'gift',
    'selected',
    'guaranteed',
    'lottery'

}
```

```python
def detect_spam(email):

    email = email.lower()

    email = re.sub(r'http\S+', '', email)

    email = re.sub(r'[^a-zA-Z\s]', '', email)

    words = email.split()

    filtered_words = []

    for word in words:

        if word not in stop_words:

            filtered_words.append(word)

    spam_count = 0

    for word in filtered_words:

        if word in spam_words:

            spam_count += 1

    if spam_count >= 2:

        return "Spam"

    else:

        return "Ham"
```

```python
df['Prediction'] = df['text'].apply(detect_spam)
```

```python
print(df[['text', 'Prediction']].head())
print("\nSummary")
print(df['Prediction'].value_counts())
```

```python
email = """
Congratulations!
You have won $5000.
Click here to claim your reward.
"""
result = detect_spam(email)
print("\nPrediction:", result)
```
